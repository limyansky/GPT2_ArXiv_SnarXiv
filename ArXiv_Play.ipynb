```python
#@title Please select a category.{ run: "auto" }
category = 'Astrophysics' #@param["Astrophysics", "Condensed Matter", "High Energy Physics - Experiment", "Tissues and Organs"]

# Data
tissue_real = ['Modeling Tumor Angiogenesis with Cellular Automata', 'Holocrine Secretion and Kino Flow in Angiosperms Their Role and Physiological Advantages in Plant Defence Mechanisms', 'The scaling of goals via homeostasis an evolutionary simulation experiment and analysis', 'Towards a Multiphysics Model for Tumor Response to Combined-Hyperthermia-Radiotherapy Treatment', 'Waves on Reissners membrane a mechanism for the propagation of otoacoustic emissions from the cochlea', 'Physics Approaches to the Spatial Distribution of Immune Cells in Tumors', 'Modelling the anabolic response of bone using a cell population model', 'Estimating the EMG response exclusively to fatigue during sustained static maximum voluntary contraction', 'Blood Glucose Measurement Based on Infra-Red Spectroscopy', 'Important factors for cell-membrane permeabilization by gold nanoparticles activated by nanosecond-laser irradiation', 'Radiomics and artificial intelligence analysis of CT data for the identification of prognostic features in multiple myeloma', 'Perspective Melanoma diagnosis and monitoring Sunrise for melanoma therapy but early detection remains in the shade', 'Topological Methods for Characterising Spatial Networks A Case Study in Tumour Vasculature', 'The OPFOS microscopy family High-resolution optical-sectioning of biomedical specimens', 'Cancer Cell Motility Optimizing Spatial Search Strategies', 'Head-to-nerve analysis of electromechanical impairments of diffuse axonal injury', 'Mimetization of the elastic properties of cancellous bone via a parameterized cellular material', 'Testing a patient-specific in-silico model to noninvasively estimate central blood pressure', 'When does humoral memory enhance infection', 'Using mathematical modelling to identify data requirements for increased prediction accuracy in radiotherapy', 'Targeting glutamate metabolism in melanoma', 'Impaired coronary blood flow at higher heart rates during atrial fibrillation investigation via multiscale modelling', 'Quantifying the impact of tissue metabolism on solute transport in feto-placental microvascular networks', 'Energy landscape analysis of cardiac fibrillation wave dynamics using pairwise maximum entropy model', 'Tracking tumor evolution via the prostate marker PSA An individual post-operative study', 'Diagnosis and Analysis of Celiac Disease and Environmental Enteropathy on Biopsy Images using Deep Learning Approaches', 'Detection of Cancer Stages via Fractal Dimension Analysis of Optical Transmission Imaging of Tissue Micro Arrays TMA', 'Morphogen Gradient from a Noisy Source', 'Dynamical anchoring of distant Arrhythmia Sources by Fibrotic Regions via Restructuring of the Activation Pattern', 'Potential Phytoextraction with in-vitro regenerated plantlets of Brassica juncea L Czern in presence of CdCl2 Cadmium accumulation and physiological parameter measurement', 'The Configuration of the Perivascular System Transporting Macromolecules in the CNS PREPRINT', 'Evidence for an unfolded border-collision bifurcation in paced cardiac tissue', 'Collective stresses drive competition between monolayers of normal and Ras-transformed cells', 'Phenomenological analysis of simple ion channel block in large populations of uncoupled cardiomyocytes', 'CXCR6 a Newly Defined Biomarker of Tissue-Specific Stem Cell Asymmetric Self-Renewal Identifies More Aggressive Human Melanoma Cancer Stem Cells', 'Decentralized physiology and the molecular basis of social life in eusocial insects', 'Cell damage and mitigation in Swiss albino mice experiment and modelling', 'Detecting Temporal shape changes with the Euler Characteristic Transform', 'Large variability in dynamical transitions in biological systems with quenched disorder', 'Quantifying Hidden Architectural Patterns in Metaplastic Tumors by Calculating the Quadrant-Slope Index QSI', 'Thermodynamic Mechanism of Life and Aging', 'Maximum Tolerated Dose Versus Metronomic Scheduling in the Treatment of Metastatic Cancers', 'Detecting Partial Rosettes in Tumor Histopathology Using The Cross Product', 'Are physiological oscillations physiological', 'A Computational Model of Levodopa-Induced Toxicity in Substantia Nigra Pars Compacta in Parkinsons Disease', 'A hybrid mathematical model for self-organizing cell migration in the zebrafish lateral line', 'Mathematical modeling of anti-tumor virus therapy Regimes with complete recovery within the framework of deterministic models', 'Fluid mechanics of mosaic ciliated tissues', 'Poroelasticity of biopolymer networks during compression theory and experiment', 'Mathematical Model of Volume Kinetics and Renal Function after Burn Injury and Resuscitation', 'Tuning gastropod locomotion Modeling the influence of mucus rheology on the cost of crawling', 'Human ovarian reserve from conception to the menopause', 'Are leaves optimally designed for self-support An investigation on giant monocots', 'Induction in myeloid leukemic cells of genes that are expressed in different normal tissues', 'Microbiome and metabolome insights into the role of the gastrointestinal-brain axis in neurodegenerative diseases unveiling potential therapeutic targets', 'Automatic Tissue Segmentation with Deep Learning in Patients with Congenital or Acquired Distortion of Brain Anatomy', 'Metabolomic response of osteosarcoma cells to nanographene oxide-mediated hyperthermia', 'A ptototype compartmental model of blood pressure distribution', 'New Universal Theory of Injury Prediction and Prevention', 'Modelling the emergence of phenotypic heterogeneity in vascularised tumours', 'Patient-specific predictions of aneurysm growth and remodeling in the ascending thoracic aorta using the homogenized constrained mixture model', 'The Universality of Cancer', 'A pseudo active kinematic constraint for a biological living soft tissue an effect of the collagen network', 'Decreased aneurysmal subarachnoid hemorrhage incidence rate in elderly population than in middle aged population a retrospective analysis of 8144 cases in Mainland China', 'Geometrical approach to tumor growth', 'Tissue evolution Mechanical interplay of adhesion pressure and heterogeneity', 'A biomimetic kidney tubule model', 'Oxygen depletion hypothesis remains controversial a mathematical model of oxygen depletion during FLASH radiation', 'Review of the Essential Roles of SMCs in ATAA Biomechanics', 'A multiscale mathematical model of cancer and its use in analyzing irradiation therapies', 'Machine learning methods for locating re-entrant drivers from electrograms in a model of atrial fibrillation', 'Effects of space structure and combination therapies on phenotypic heterogeneity and drug resistance in solid tumors', 'Olive oil by-product as functional ingredient in bakery products', 'A tool for user friendly cloud based whole slide image segmentation', 'A fully resolved active musculo-mechanical model for esophageal transport', 'Critical fluctuations of the electrical activity of the heart Shortcomings of models of excitability and interpretation', 'Chronic stress may disrupt covariant fluctuations of vitamin D and cortisol plasma levels in pregnant sheep during the last trimester a preliminary report', 'Emergence of Anti-Cancer Drug Resistance Exploring the Importance of the Microenvironmental Niche via a Spatial Model', 'Finite element methods for large-strain poroelasticitychemotaxis models simulating the formation of myocardial oedema', 'Simulating the time series of a selected gene expression profile in an agent-based tumor model', 'Using deceased-donor kidneys to initiate chains of living donor kidney paired donations algorithms and experimentation', 'Living Tissue Self-Regulation as a Self-Organization Phenomenon', 'OCTAVA an open-source toolbox for quantitative analysis of optical coherence tomography angiography images', 'Canine Olfactory Differentiation of Cancer A Review of the Literature', 'A hexamer origin of the echinoderms five rays', 'Automated Coronary Arteries Labeling Via Geometric Deep Learning', 'Emergence of proto-organisms from bistable stochastic differentiation and adhesion', 'Levy model of cancer', 'Algebraic framework for determining laminar pattern bifurcations by lateral-inhibition in 2D and 3D bilayer geometries', 'Femoral fracture load and damage localisation pattern prediction based on a quasi-brittle law linear and non-linear FE meshing', 'Space nutrition the key role of nutrition in human space flight', 'Similarities vs key discrepancy between tuberculosis and cancer', 'Extending the Relative Seriality Formalism for Interpretable Deep Learning of Normal Tissue Complication Probability Models', 'When the clock strikes Modeling the relation between circadian rhythms and cardiac arrhythmias', 'Mathematical Properties of Pump-Leak Models of Cell Volume Control and Electrolyte Balance', 'Consequences du changement climatique pour les maladies a transmission vectorielle et impact en assurance de personnes', 'Spatio-temporal Models of Lymphangiogenesis in Wound Healing', 'Decreased serum vitamin D level as a prognostic marker in patients with COVID-19', 'A Poroelastic Mixture Model of Mechanobiological Processes in Tissue Engineering Part II Numerical Simulations', 'The Cancer Diaspora Metastasis beyond the seed and soil hypothesis']
condensed_real = ['Single photon detection performance of highly disordered NbTiN thin films', 'Interparticle interaction and size effect in polymer coated magnetite nanoparticles', 'Robust formation of nanoscale magnetic skyrmions in easy-plane thin film multilayers with low damping', 'Femtosecond laser driven molecular dynamics on surfaces Photodesorption of molecular oxygen from Ag110', 'Adaptive hard and tough mechanical response in single-crystal B1 VNx ceramics via control of anion vacancies', 'Band Structure of ABC-Stacked Graphene Trilayers', 'Antiferromagnetic spin-12 chains in NOCuNO33 a microscopic study', 'Strong Electron-Phonon Coupling Superconductivity Induced by a Low-lying Phonon in IrGe', 'Laser-induced fluorescence based characterization method for aggregation behavior of rhodamine B RhB in water ethanol and propanol', 'Zitterbewegung and its effects on electrons in semiconductors', 'Magnetization switching in polycrystalline Mn3Sn thin film induced by self-generated spin-polarized current', 'Synergistic Photon Management and Strain-Induced Band Gap Engineering of Two-Dimensional MoS2 Using Semimetal Composite Nanostructures', 'Molecular Configuration Around Single Vacancy in Solid CO2 at T 0 K Studied by Monte Carlo Simulation Technique', 'Ab initio studies of ionization potentials of hydrated hydroxide and hydronium', 'Determining sample alignment in X-ray Reflectometry using thickness and density from GaAsAlAs multilayer certified reference materials', 'Interfacial contribution to the dielectric response in semiconducting LaBiMn43Co23O6', 'On-site approximation for spin-orbit coupling in LCAO density functional methods', 'Linear response time-dependent density functional theory of the Hubbard dimer', 'Electronic and optical properties of Mn impurities in ultra-thin ZnO nanowires insights from density-functional theory', 'Scaling law for the onset of the surface wrinkling of multilayer tubes', 'Charge transfer at hybrid inorganic-organic interfaces', 'Co-existence of Topological Non-trivial and Spin Gapless Semiconducting Behavior in MnPO4 A Composite Quantum Compound', 'Superconductivity above 80 K in polyhydrides of hafnium', 'Surface phase transitions in BiFeO3 below room temperature', 'Effect of secondary LuNiSn phase on thermoelectric properties of half-Heusler alloy LuNiSb', 'Jamming transition in non-spherical particle systems pentagons vs disks', 'The electronic structure of zircon-type orthovanadates Effects of high-pressure and cation substitution', 'Origins of GaN0 0 0 1 surface reconstructions', 'Spin-spin interactions in solids from mixed all-electron and pseudopotential calculations - a path to screening materials for spin qubits', 'Ultra-high mechanical stretchability and controllable topological phase transitions in two-dimensional arsenic', 'Probing optical excitations in chevron-like armchair graphene nanoribbons', 'Thermally activated switching of perpendicular magnet by spin-orbit spin torque', 'Lattice Dynamics and Thermal Equation of State of Platinum', 'Chiral Bobbers and Skyrmions in Epitaxial FeGeSi111 Films', 'Effect of External Normal and Parallel Electric Fields on 180o Ferroelectric Domain Walls in PbTiO3', 'Phosphorene oxides bandgap engineering of phosphorene by oxidation', 'Fabrication and properties of luminescence polymer composites with erbiumytterbium oxides and gold nanoparticles', 'Temperature Effects on the Fracture Dynamics and Elastic Properties of Popgraphene Membranes', 'Exploring the best scenario for understanding the high temperature thermoelectric behaviour of Fe2VAl', 'Unusual plastic deformation and damage features in Titanium experimental tests and constitutive modeling', 'The influence of non-idealities on the thermoelectric power factor of nanostructured superlattices', 'Critical pore radius and transport properties of disordered hard- and overlapping-sphere models', 'Chemical bonding in carbide MXene nanosheets', 'Magnetic interactions of supported magnetic clusters', 'Chemical Influences on Ice Crystal Growth from Vapor', 'Nonmonotonic photostability of BA2MAn-1PbnI3n1 homologous layered perovskites', 'Ab initio study on the effects of transition metal doping of Mg2NiH4', 'Unraveling Energetic Disorder in Organic Bulk Heterojunction Photovoltaics by Capacitance-Voltage Spectroscopy', 'Colossal piezoresistance in narrow-gap Eu5In2Sb6', 'Nearly total spin polarization in La23Sr13MnO3 from tunneling experiments', 'Short range ordering of heavy element columns in nickel based superalloys', 'A Review of Sc-containing Scandate Thermionic Cathodes', 'Synthesis and Bulk Properties of Oxychloride Superconductor Ca2-xNaxCuO2Cl2', 'Thermal conductance of graphene and dimerite', 'Boundary Conditions in an Electric Current Contact', 'A Dynamic Atomistic-Continuum Method for the Simulation of Crystalline Materials', 'Magnetic domain wall motion in SrRuO3 thin films', 'Rashba-induced spin texture and spin-layer-locking effects in the antiferromagnetic CrI3 bilayer', 'Contributions of Co and Fe orbitals to Perpendicular Magnetic Anisotropy of MgOCoFeB Bilayers with Spin-Orbit-Torque-Related Ta W IrMn and Ti Underlayers', 'Stability of magnetic nanoparticles inside ferromagnetic nanotubes', 'The Effect of Structural Conformational Changes on Charge Transfer States in a Light-Harvesting Carotenoid-diaryl-Porphyrin-C60 Molecular Triad', 'Static and Dynamic Oxide-Trapped-Charge-Induced Variability in Nanoscale CMOS Circuits', 'In-fiber second-harmonic generation with embedded two-dimensional materials', 'Exact analytical solution for electrostatic field produced by biased atomic force microscope tip dwelling above dielectric-conductor bi-layer', 'Magnetically inducedenhanced coarsening in thin films', 'Casimir interactions in graphene systems', 'Persistent photovoltage in methylammonium lead iodide perovskite solar cells', 'Reversible Structural Transition in Epitaxial Manganite Film', 'Study on the hydrogen behavior in Y and Co doped barium-zirconate', 'Correlated Doping in Semiconductors The Role of Donors in III-V Diluted Magnetic Semiconductors', 'Auto-oscillation threshold narrow spectral lines and line jitter in spin-torque oscillators based on MgO magnetic tunnel junctions', 'Unified paradigm for interface dynamics', 'Properties of patchy colloidal particles close to a surface a Monte Carlo and density functional study', 'Temperature dependence of the energy band gap in ZrTe5 implications for the topological phase', 'Optical conductivity due to orbital polarons in systems with orbital degeneracy', 'New Families of Carbon Nanotubes', 'Single-atom catalysts boost nitrogen electroreduction reaction', 'Configurational electronic entropy and the phase diagram of mixed-valence oxides the case of LixFePO4', 'Computing total energies in complex materials using charge self-consistent DFTDMFT', 'Quasi-two-dimensional Dirac fermions and quantum magnetoresistance in LaAgBi2', 'In situ and non situ surface x-ray scattering studies of in plane Cs ordering in Helmholtz planes of Pt111 surface', 'Symmetry analysis of k 13130 three-sublattice ordering in A14B51 compounds', 'Quantization Phenomena of critical Hamiltonians in 2D systems', 'Growth and Thermo-driven Crystalline Phase Transition of Metastable Monolayer 1T-WSe2 Thin Film', 'Synthesis crystal structure and magnetic properties of RE3Sb3Mg2O14 RELa Pr Sm Eu Tb Ho new 2D Kagome materials', 'Field Effect Transistor Based on KTaO3 Perovskite', 'Stimulated emission from ZnO thin films with high optical gain and low loss', 'Direct imaging of local atomic structures in zeolite using novel low-dose scanning transmission electron microscopy', 'Weak localization and electron-electron interactions in Indium-doped ZnO nanowires', 'Prediction of Stable Ground-State Uranium Nitrides at Ambient and High Pressures', 'Phase-change silicon as an ultrafast active photonic platform', 'Evidence for Incipient Ferroelectricity in YCrO3', 'A many-body perturbation theory approach to the electron-phonon interaction with density-functional theory as a starting point', 'Electric conductivity of lead iron niobate', 'Nonmonotonic band gap evolution in bent phosphorene nanosheets', 'Ion beam shaping and downsizing of nanostructures', 'The Normal State Resistivity of Grain Boundaries in YBa2Cu3O7-delta', 'Theory of interacting dislocations on cylinders', 'Structural and electronic properties of highly doped topological insulator Bi2Se3 crystals', 'Band Bending Independent of Surface Passivation in ZnOCdSCuInGaSSe2 Heterojunctions and CrCuInGaSSe2 Schottky Contacts']
astro_real = ['Resonant micro-instabilities at quasi-parallel collisionless shocks cause or consequence of shock reformation', 'The Optical Luminosity Function of Gamma-ray Bursts deduced from ROTSE-III Observations', 'Long-term XMM-Newton view of magnetar CXOU J0100431-721134 Comprehensive spectral and temporal results', 'Pulsars and Gravitational Waves', 'Pulsars are cool Seriously', 'PIC Simulations of Velocity-Space Instabilities in a Decreasing Magnetic Field Viscosity and Thermal Conduction', 'Ultraviolet Observations of Super-Chandrasekhar Mass Type Ia Supernova Candidates with Swift UVOT', 'Astro 2020 Science White Paper Joint Gravitational Wave and Electromagnetic Astronomy with LIGO and LSST in the 2020s', 'Angular Momentum Transport by Acoustic Modes Generated in the Boundary Layer II MHD Simulations', 'Probing the disk-corona systems and broad line regions of changing-look quasars with X-ray and optical observations', 'Radio Remnants of Compact Binary Mergers - the Electromagnetic Signal that will follow the Gravitational Waves', 'Highly relativistic spin-gravity coupling for fermions', 'Simulating Cherenkov Telescope Array observation of RX J17137-3946', 'Extragalactic Very-High-Energy gamma-ray background', 'Discovery of a hot ultramassive rapidly rotating DBA White Dwarf', 'Reassessing candidate eccentric binary black holes Results with a model including higher-order modes', 'A noise independent method of measuring sky location and chirp mass of gravitational waves single sources using non-simultaneously timing of pulsars', 'Obscuration by Gas and Dust in Luminous Quasars', 'A New Flaring Black Widow Candidate and Demographics of Black Widow Millisecond Pulsars in the Galactic Field', 'Temporal and spectral evolution in X- and gamma-rays of magnetar 1E 15470-5408 since its October 2008 outburst the discovery of a transient hard pulsed component after its January 2009 outburst', 'Multi-timescale analysis of phase transitions in precessing black-hole binaries', 'Common Envelope Wind Tunnel Range of Applicability and Self-Similarity in Realistic Stellar Envelopes', 'Astrophysical Fluids of Novae High Resolution Pre-decay X-ray spectrum of V4743 Sagittarii', 'The cycle of metals in the infalling elliptical galaxy NGC 1404', 'Deep observation of the NGC 1275 region with MAGIC search of diffuse gamma-ray emission from cosmic rays in the Perseus cluster', 'Neutrino emission from FRB-emitting magnetars', 'Frequency-Dependent Core Shifts in Ultracompact Quasars', 'The Equatorial Current Sheet and other interesting features of the Pulsar Magnetosphere', 'Nucleosynthesis in 2D Core-Collapse Supernovae of 112 and 170 Modot Progenitors Implications for Mo and Ru Production', 'SN 2020jfo A short plateau Type II supernova from a low mass progenitor', 'A power-law break in the near-infrared power spectrum of the Galactic center black hole', 'Detecting Fast Radio Bursts in the Milky Way', 'Energy Dependence of Direct Detection Cross Section for Asymmetric Mirror Dark Matter', 'The Effect of Mission Duration on LISA Science Objectives', 'Strong outburst activity of the X-ray pulsar X Persei during 2001-2011', 'Air showers and hadronic interactions with CORSIKA 8', 'Population synthesis of AX J17456-2901 X-ray nova type binaries with rapidly decreasing orbital periods', 'Superorbital modulation at GeV energies in the gamma-ray binary LS I 61 303', 'Analyses on the Variability Asymmetry of Kepler AGNs', 'Four Years of Fermi LAT Observations of Narrow-Line Seyfert 1 Galaxies', 'Interpretation of neutrino flux limits from neutrino telescopes on the Hillas plot', 'Are T Tauri stars gamma-ray emitters', 'Neutron star stiff equation of state derived from cooling phases of the X-ray burster 4U 1724-307', 'Search for Quantum Gravity Using Astrophysical Neutrino Flavour with IceCube', 'A detailed study of the modulation of the optical light from Sk160SMC X-1', 'Broadband Spectral Fitting of Blazars using XSPEC', 'The Stokes Phase Portraits of Descattered Pulse Profiles of a Few Pulsars', 'One large glitch in PSR B1737-30 detected with the TMRT', 'Application of a symplectic integrator in a non-integrable relativistic system', 'The Stability of Core-Envelope Models for the Crab and the Vela pulsars', 'Wiggly tails a gravitational wave signature of massive fields around black holes', 'Dark Matter Search at Colliders and Neutrino Floor', 'Voronoi chords in flat cosmology', 'A simple analytic model for astrophysical S-factors', 'Informing the Cataclysmic Variable Donor Sequence from Gaia DR2 Color-Magnitude and Inferred Variability Metrics', 'Hall Scrambling on Black Hole Horizons', 'Modeling Reconstructed Images of Jets Launched by SANE Super-Eddington Accretion Flows Around SMBHs with the ngEHT', 'Extreme Gravity Tests with Gravitational Waves from Compact Binary Coalescences I Inspiral-Merger', 'Stability windows at finite temperature', 'Distributions of Gamma-Ray Bursts and Blazars in the Lrm p-Erm p Plane and Possible Implications for their Radiation Physics', 'Photonic chiral vortical effect', 'SN 2009ip and SN 2010mc as dual-shock Quark-Novae', 'The relativistic kinetic dispersion relation Comparison of the relativistic Bhatnagar-Gross-Krook model and Grads 14-moment expansion', 'Can CoGeNT and DAMA Modulations Be Due to Dark Matter', 'The difficulty of inferring progenitor masses from Type II-Plateau supernova light curves', 'A Hard-to-Soft State Transition of Aquila X-1 Observed with Suzaku', 'The SUrvey for Pulsars and Extragalactic Radio Bursts I Survey Description and Overview', 'The Polarization Properties of Inverse Compton Emission and Implications for Blazar Observations with the GEMS X-Ray Polarimeter', 'Detailed properties of gravitational-wave mergers from flyby perturbations of wide binary black holes in the field', 'Deep XMM-Newton Spectroscopic and Timing Observations of the Isolated Radio Millisecond Pulsar PSR J00300451', 'TeV cosmic-ray proton and helium spectra in the myriad model', 'The Diversity of Transients from Magnetar Birth', 'Toward Realistic Models of Core Collapse Supernovae A Brief Review', 'Evolution of X-ray spectra and light curves of V1494 Aquilae', 'Surveying the Dynamic Radio Sky with the Long Wavelength Demonstrator Array', 'Vertical Structure of the Outer Accretion Disk in Persistent Low-Mass X-Ray Binaries', 'Multipolar electromagnetic fields around neutron stars general-relativistic vacuum solutions', 'Low frequency view of GW 170817GRB 170817A with the Giant Meterwave Radio Telescope', 'Binary Neutron Star Mergers Mass Ejection Electromagnetic Counterparts and Nucleosynthesis', 'Cosmic-Ray Acceleration Efficiency vs Temperature Equilibration the Case of SNR 0509-675', 'Gravitational waves or X-ray counterpart No need to choose', 'Comparing simulated 26Al maps to gamma-ray measurements', 'Ensemble Fluctuations of the Flux and Nuclear Composition of Ultra-High Energy Cosmic Ray Nuclei', 'tt bajes Bayesian inference of multimessenger astrophysical data methods and application to gravitational-waves', 'Spectral and polarization study of the double relics in Abell 3376 using the GMRT and the VLA', 'A Mid-infrared Study of Superluminous Supernovae', 'Recent pulsar results from VERITAS on Geminga and the missing link binary pulsar PSR J10230038', 'A gamma-ray periodic modulation in Globular Cluster 47 Tucanae', 'IC 751 a new changing-look AGN discovered by NuSTAR', 'SN 2017gci a nearby Type I Superluminous Supernova with a bumpy tail', 'Testing the Kerr black hole hypothesis with GRS 1716-249 by combining the continuum-fitting and the iron-line methods', 'Choked jet model for the neutrino emission associated with Tidal Disruption Events', 'The Fermi Gamma-Ray Haze from Dark Matter Annihilations and Anisotropic Diffusion', 'Effect of acceleration and escape of energetic particles on spectral steepening at shocks', 'Has LIGO Detected Primordial Black Hole Dark Matter -- Tidal Disruption in Binary Black Hole Formation', 'Multi-wavelength Campaign on NGC 7469 VI Photoionisation Modelling of the Emission Line Regions and the Warm Absorber', 'Supernova-scope for the Direct Search of Supernova Axions', 'News from Cosmic Ray Air Showers ICRC 2019 -- Cosmic Ray Indirect Rapport', 'Angular momentum redistribution by SASI spiral modes and consequences for neutron star spins', 'Full-3D relativistic MHD simulations of Bow Shock Pulsar Wind Nebulae dynamics']
hepEX_real = ['Multiboson production in W decays', 'NLONLL squark and gluino production cross-sections with threshold-improved parton distributions', 'Interpretation of Yb 10750 as a tetraquark and its production mechanism', 'A Light Higgs at the LHC and the B-Anomalies', 'Empirical parameterization of the K- - pi- pi0 pi0 decay Dalitz plot', 'Jet-like correlations with respect to K0rm S and Lambda barLambda in pp and Pb-Pb collisions at mathbfitsqrtsmathrmNN 502 TeV', 'Neutrinoless Double Beta Decay Present and Future', 'Dispersion Relation Constrained Partial Wave Analysis of piN Elastic and piNtoetaN Scattering Data The Baryon Spectrum', 'Heavy-Flavour Production in Pb-Pb collisions at the LHC with ALICE', 'Leading jets and energy loss', 'Measurement of the neutrino velocity with the ICARUS detector at the CNGS beam', 'New Results on Tau Lepton', 'Search for b--u transitions in B- -- DK- and B- -- DK- Decays', 'Separation of a single photon and products of the pi0eta K0s meson neutral decay channels in the CMS electromagnetic calorimeter using neural network', 'Science with e-ASTROGAM A space mission for MeV-GeV gamma-ray astrophysics', 'Proposal for chiral bosons search at LHC via their unique new signature', 'Search for Leptoquarks', 'What about a beta-beam facility for low energy neutrinos', 'Study of the breakdown voltage of SiPMs', 'Anomalous Gauge Interactions of the Higgs Boson Precision Constraints and Weak Boson Scatterings', 'New results on semileptonic B decays from the Belle experiment', 'Quasi Anomalous Knowledge Searching for new physics with embedded knowledge', 'Search for the Rare Decays JPsi -- Ds- e nue JPsi -- D- e nue and JPsi -- D0bar e e-', 'Improved Measurements of barB0 to D0 barK0 and barB0 to D0 barK0 Branching Fractions', 'Bounds on heavy Majorana neutrinos in type-I seesaw and implications for collider searches', 'Hadron Physics at J-PARC', 'Implementation of the Dead-Time Free F1 TDC in the COMPASS Detector Readout', 'The Calculation of Sparticle and Higgs Decays in the Minimal and Next-to-Minimal Supersymmetric Standard Models SOFTSUSY40', 'Measurements of the Sigmac0 and Sigmac Mass Splittings', 'Heavy Baryon-Antibaryon Molecules in Effective Field Theory', 'Precision Electroweak Measurements at Run 2 and Beyond', 'Dynamical origin and the pole structure of X3872', 'Jet Cross-Sections and alphas in DIS at HERA', 'Low energy reaction K-prightarrowLambdaeta and the negative parity Lambda resonances', 'Search for supersymmetry in pp collisions at sqrts 13 TeV with 137 fb-1 in final states with a single lepton using the sum of masses of large-radius jets', 'Beautiful and Charming Energy Correlators', 'Distinguishing between MSSM and NMSSM through Delta F2 processes', 'CompAZ parametrization of the luminosity spectra for the photon collider', 'Tests on NaITl crystals for WIMP search at the Yangyang Underground Laboratory', 'Measurement of CP asymmetry in Cabibbo suppressed D0 decays', 'Studying Transverse Momentum Dependent Distributions in Polarized Proton Collisions Via Azimuthal Single Spin Asymmetries of Charged Pions in Jets', 'Parity Violation and Chiral Oscillation of Cosmological Relic Neutrinos', 'Search for D0 - ellell- decays and for CP violation in Ds to KS0pi and Ds to KS0K at BELLE', 'Study of charging up effect in a triple GEM detector', 'S-wave K-pi System in Dto K-pipi Decays from Fermilab E791', 'Effect of supersymmetric phases on lepton dipole moments and rare lepton decays', 'Experimental results with boosted top quarks in the final state', 'Magnetic moments of doubly heavy baryons in light-cone QCD', 'Direct photon production at next-to-next-to-leading order', 'Direct detection of Higgs-portal dark matter at the LHC', 'Scattering studies with low-energy kaon-proton femtoscopy in proton-proton collisions at the LHC', 'Single-hit resolution measurement with MEG II drift chamber prototypes', 'Constraining new physics with collider measurements of Standard Model signatures', 'Mapping the genuine bosonic quartic couplings', 'Jpsi polarization at Tevatron and LHC Nonrelativistic-QCD factorization at the crossroads', 'Search for supersymmetry with razor variables in pp collisions at sqrts 7 TeV', 'Observation of a structure at 184 GeVc2 in the 3pipi- mass spectrum in Jpsirightarrow gamma 3pipi- decays', 'Observation of radiative B0 - phi K0 gamma decays', 'Beauty-full Tetraquarks', 'Mass spectra of mesons containing charm quarks - continuum limit results from twisted mass fermions', 'Decays of the Upsilon1S into a Photon and Two Charged Hadrons', 'Highlights on Supersymmetry and Exotic Searches at the LHC', 'Energetic neutron identification with pulse shape discrimination in pure CsI crystals', 'Doubly Heavy Baryons Expanded in 1mQ', 'Origin of Constrained Maximal CP Violation in Flavor Symmetry', 'Production of charged pions kaons and protons in ee- annihilations into hadrons at sqrts 1054 GeV', 'Test of lepton universality with B0 rightarrow K0ellell- decays', 'Search for charginos and neutralinos in final states with two boosted hadronically decaying bosons and missing transverse momentum in pp collisions at sqrts13 TeV with the ATLAS detector', 'Search for lepton-number-violating signals in the charm sector', 'First data from the CUPID-Mo neutrinoless double beta decay experiment', 'The bound state problem of S-wave heavy quark meson-aitimeson systems', 'Comment on the review paper of Amsler', 'Probing BFKL Dynamics in the Dijet Cross Section at Large Rapidity Intervals in ppbar Collisions at sqrts1800 and 630 GeV', 'Measurement of the Resonance Parameters of the chi113P1 and chi213P2 States of Charmonium formed in Antiproton-Proton Annihilations', 'The pi0 eta eta - gamma gammaQ2 Decay Rates and Radii', 'Heterogeneous High Throughput Scientific Computing with APM X-Gene and Intel Xeon Phi', 'Precision physics with pile-up insensitive observables', 'Influence of correlations between yields on the chemical freeze-out temperature', 'New calculations of neutron and kaon decays', 'First Dark Matter Search Results from the LUX-ZEPLIN LZ Experiment', 'Structure of heavy baryons in a pion mean-field approach', 'Observation of the Semileptonic Decay D0 to a0980- e nue and Evidence for D to a09800 e nue', 'A Search for Nucleon Decay via n rightarrow barnu pi0 and p rightarrow barnu pi in Super-Kamiokande', 'Breaking down the entire spectrum of spin correlations of a pair of particles involving fermions and gauge bosons', 'Inference of neutrino nature and Majorana CP phases from 0nubetabeta decays with inverted mass ordering', 'Elliptic Flow from a Transversally Thermalized Fireball', 'Research and Development towards a Detector for a High Energy Electron-Positron Linear Collider', 'A Search for Excited Fermions in e p Collisions at HERA', 'Measurement of the CKM angle gamma phi3 using B-DK decays', 'Digital radio detection of cosmic rays achievements status and perspectives', 'First Measurements of the Exclusive Decays of the Upsilon5S to B Meson Final States and Improved Bs Mass Measurement', 'Null radiation zone at the LHC', 'The axion dark matter echo a detailed analysis', 'Direct detection of non-galactic light dark matter', 'Bruno Pontecorvo and neutrino physics', 'Heavy Quarkonium Production at Threshold from JLab to EIC', 'CP Violation in B0 decays to Charmonium and Charm Final States at BaBar', 'Quark flavour observables in the Littlest Higgs model with T-parity after LHC Run 1', 'The CLEO-III RICH Detector and Beam Test Results', 'Search for hcrightarrowpipi-Jpsi via psi3686rightarrow pi0pipi-Jpsi']

tissue_fake = ['A computational model of neural activity in mammalian brainstem structures', 'Research in molecular dynamics', 'A model for the growth and evolution of neural tube models of bone growth', 'Atrial Fibrillation Associated with Multiple Radiotherapy in Cancers', 'It is known that when induced into HBRs during photothermal stress a photospheric surface morphogenesis model is created for a therapeutic dose of cancer', 'An overview of the biology and molecular dynamics of caspase-induced apoptosis in cytosol', 'The role of dietary factors in the clinical development and maintenance of neuropathic pain in patients with terminal spinal stenosis', 'The mechanistic processes underlying functional changes in the human respiratory system', 'Information on bone density in animals', 'The genetic models for cancer susceptibility and susceptibility to oral cancer epithelial cell carcinoma', ' Understanding Biophysics of Naturism and Nanotubes The Simulation of Accelerating the Synthesis of Hydrogen and Carbon from Carbone Fibers (Sideways Tissue) to Maturation and Thiolation in a Systemic Tumor - Synthetic Bioinspired Engineering for Pancreatic Cancer', 'A multilevel linear model for the identification of microenvironmental risk factors for glioma', ' Phase-dependent response to hypoxic stress on the cerebral blood supply', 'An overview of the developmental development and dynamics of an oral cavity and mucosal barrier in an experimental model', 'The stability and instability of protein and protein complexes in the development of disease induced maculopods', 'Tissue cell populations in a tumor model of metastatic sarcomas', 'A Multicellular Response-Dependent Membrane-mediated Cell Flow and Signaling', 'A mechanistic model of cell invasion and its role in cell death in hepatocytes-derived hepatomas', 'Spatial modeling and hierarchical networks for prediction and prediction of tumors', 'The neural network of cerebral blood flow around ventricular chondrocytes', 'The interaction between dietary fiber and plasma lipoprotein cholesterol and growth in relation growth and disease', ' PhaseDynamics in Lymphomatoma', 'Lateral lobe and orbitofrontal cortex anatomy', 'A low-resolution model of human osteoporosis', 'The rate of death from respiratory failure during acute administration of anhydroxyluridectomy', 'The effect of sodium chlorathrin on polyphenol metabolism and bioavailability of monounsaturated fatty acids in a small intestine', 'A multivariate network of models for predicting health-related cancer cell models', 'Use of an internal thermodynamic model to assess body weight induced by fasting and nonfasting tissue', 'As the function of oxidative stress activates the transport of plasma proteins to the', 'A quantitative estimation of the metabolic activity of an embryo by imaging spectroscopy', 'A framework for measuring pain levels from functional tissue tomography', 'Overexpressive activity of a high-intensity stimulation matrix', 'The functional role of the brain in regulating cognition of schizophrenia', 'The dynamics of protein utilization during a high energy cell adhesion cycle and tissue growth during stress and inflammation', 'Is molecular and cellular aging caused by tumor necrosis? A review of the molecular nature of metastasis with a focus on a disease and its relationships with biomarkers and target parameters', 'Data on microstructures and nanostructures', 'The efficacy and safety of a novel probiotic food treatment for children with developmental disorders', 'In the long term, the pathogenesis of cerebral atrophy', '-phase transition dynamics of cytokine in the lymphatic nervous system', '(Ascorpion-based model of vascular infassemble) Modeling osteoporosis with an ultra-low-volume model', 'A model for developmental analysis of human immunodeficiency virus infections', 'A novel pathogenic model of breast cancer with implications for the development of cancer', ' Application and Treatment of Sulfuric Acid Inhalation in Doses', 'Bummed and enlarged lymphatic process', 'Data analysis analysis of growth in vivo melanoma lesions with multiplexing', 'Multi-dimensional lattice modeling for rapid neuronal growth', 'Is the formation of a high-frequency transverse adenomatopoiesis syndrome', 'New insights into neural networks with the Cauchyneocyte and Cholecystis macromolecules', 'A theoretical framework for the application of statistical modeling to biological models of lymphoblastic infection on a human model', 'The interaction of low-density lipoprotein and non-coagulating tissue with muscle mass and lipids and an increase in T1-mediated cellular death in a dynamic model of human cardiac fibrosis', 'A quantitative tomography model for cell carcinogenesis', 'Prostate Cancer Fibrosis and Immunotherapy', 'A Novel Prophylactic Mechanic for Glucose Resistant Patients With Chronic Stuttering', 'the effect of dietary intake on insulin sensitivity and pancreatic fibrosis in patients with chronic pancreatitis', 'On the role of the phthalassemia and its effect on the vascular system in human disease', 'A simple study of the emergence and survival of cellular viruses in the human respiratory and spleen during microcalculus', 'The study of human neural circuit regulation induced by induced pluripotent stem cell line', 'An Introduction to Bioprinting Optimal Optic Excitation Models for Neuromodulation', 'and method for diagnosis of hepatic cancer using the NMR spectroscopy', 'The role of multidisciplinary approaches for diagnosis of cancer with radiation therapy', 'The Effect of Modulation of the Cardiac Cytokine Morphogenesis and The Structure of Circulatory Circuit Systems', 'At a Glance At Heart-Related Developmental Variability in Heart Dynamics By Gene Flow Infection Response To The Immune Bacterial Species', 'Starch-transferase-induced vasculature-mediated relaxation-conditioning of myocardial ischemia', 'Vivo-electron microscopy', 'All patients with chronic liver disease', 'Multi-T-cell Recurrence Rate of Cell Growth Reversed in Vivo Biomarkers and Optimal Determination of Vascular Cells', 'The influence of estrogen on post-mortem brain tissue growth and pathology', 'A kinetic analysis of the interactions between neuronal dynamics and the pathophysiology of traumatic brain injury', 'The characterization of the plasma membrane as an integrated domain model for metastasis of primary tumors', 'The feasibility of improving clinical outcomes in a multicellular mouse model of age-related disease', 'I have demonstrated that high-density lipoproteins in relation to cellular metabolism have a positive effect on immune function', 'The effects of antioxidant enzymes on cell death from plaque decay', 'We perform a detailed review of the study of a common polymorphism in mitochondrial DNA under developmental stress', 'Is a monozygotic deletion of chromosome 7 deletion in the Cretylovus', 'A model for drug resistance and tumor growth in a cancer-resistant human lymphoid tumor', 'The neural and electrophysiological networks on an integrated scale', '(A study to determine the effect of the anticoagulation of estrogen on tissue growth and aging )', ' Biodynamic Properties of the Polymer Organometallic Hydroponic Sheet', 'It is possible to predict the formation of the hypoxia-induced hypoadrenylation in murine liver', ' Rapid and rapid-growing growth of fibroblasts from tumor tissue tissue', 'The role of calcium in regulating bone turnover in vertebrates', 'Tumor cells and tissue growth in blood', 'The Effects of Chronic Stress on Blood Supply and Blood Pressure in Chronic Prostate Cancer Patients', 'Heterogeneity of cardiac and pulmonary vessels', 'In vivo evaluation of intracellular calcium induced by low-dose sodium chloride in cardiac fibroblast fibrosis', 'A model for cardiac arrest using the Catecholamine System and the Heart Rhythm Model', 'Tissue growth rate among healthy adults and pre-menopausal women', 'Alleleling Morphinomas', 'The Effect of Anticholinergic Modulation on Cardiac Function in Human Subjects', 'the role of neuronal activity in cortical development', 'A comparative analysis of body temperature changes in the human heart after rapid coronary artery disease', 'A Novel Framework for Pathology Evaluation of Hematocrit', 'Numerical model of age-related neuronal death in the brain in young subjects A hierarchical clustering of neural tissues', 'Aspartate receptors in the brain are reduced in vivo during cell growth During the development of mice', 'For the detection of tissue tumor growth and response to chemotherapy using ultrasound', 'A Statistical Model for the Detection of Immune Transplantation', 'Efficient adaptive analysis of stress-induced tissue biogenesis in hepatocyte cultures', 'and an application of biofuels to the treatment of wounds', 'It is hypothesized that the brain develops a-scale brain anatomy perspective on the cellular structure and function of the human brain', 'A model for the effects of intraarterial airway administration on vascular permeability in patients with recurrent heart defects']
condensed_fake = ['All-electron spin oscillations at bulk FeMgO interfaces', 'Atomic-scale thermal and insulating properties of CrI2N-ZnN heterostructures from first principles', 'Perturbation induced the evolution of the lattice-field order in the superconducting metal CuSnO2', 'Surface and transport properties of carbon nanotubes', 'Ultrafast ferromagnetism and ferroelectricity in three-dimensional GaAs', 'Coherence Dependence of Band Inversion on the Effective Spin Coupling Frequency', 'The molecular dynamics and surface states in a nanocomposite crystal', 'Single layer graphene is a critical low energy thermoelectric material', 'Correlation of spin modes and low-energy X-rays from first principles', 'Graphene-Aloe-2D Photonic Device in Doped Titanium Alloy', 'First-principles study of high-pressure hydrogen injection in a SrTiO3 superconductor', 'Electron transport properties of the SrTiO3 interface', 'First Principles Approach to Multiscale Structural Phase Transition in Single Dichalcogenides', 'A quantum moment theory based on a model of the magnetoepitaxy of diamond nanoparticles', 'Magnetic and nonmagnetic anisotropy of the LaO3', 'Electronic structures of graphene bilayers with low-temperature defects', 'An Inertial Magnetic Density Functional study of a Magnetic Tunnel', 'Multiband electronic structures of nonlinear spin-orbit interactions in Sr2NiO4-xMnO3 double quantum wells', 'Numerical analysis of the effect of room temperature and temperature-dependent transition metal nitride nanoparticles on the spin-orbit transition energy in single crystal lattice materials', 'High-speed electron transport in perovskites at the edge of the domain wall', 'New Spin-orbit Coupling in Zr and Mn', 'Large-band gap in the domain of van der Waals polycrystalline solids from ab initio calculations', 'Phonon Anomaly on a Liquid Material and Nanostructure', 'Localized magnetocaloric effect in CrFe-doped CoZnO thin films studied by Raman spectroscopy', 'Strong excitonic signature in a Ni-doped ZnSn alloys', 'Observation of the antiferromagnetic phase transition in C3C3 and C4C4 films of epitaxial LaV3AlGaAs2', 'Structural and Magnetic Properties of MgO2 A First Principles Study', 'Exciton scattering in an alpha-Mn3O12', 'Self-assembly and re-centrosymmetry of the SrRuO3 thin films from an accurate computational method', 'Anisotropic FFT and Magnetic Room-Temperature Electron Emission in MoS2', 'Intra-linear thermal transport and molecular dynamics in multiferroics', 'On the theory of symmetry and its relationship to the order of the lattice effects', 'Determining the Ternary Stress Phase in Manganese Layers by Magnetic Resonance Field X-ray Thermal Scattering', 'Ab initio study of superconductivity and band gap band structure in Ba2Cu4O6 from first principles', 'On the origin of the ferroelectric gap in CuSe2', 'Density-functional theory in molecular magnets', 'Electronic structure and ferroelectricity of silicon thin film', 'Graphene Nanocavity', 'Unraveling the mechanical force due to self-assembly of intercalated solids', 'Two-Dimensional Monolayer on Carbon Nanotubes', 'Superconductivity in Cr2FeFe3 Thin Films using GaAs', 'Role of the Localized Photonics in Crystal Crystal Morphology and Magnetic Field Phase Transitions', 'Low-Phasing Graphene by Structural Transfer', 'Modeling thermodynamic effects for many-body transition metal oxides of iron', 'Phase-field analysis of a nanoscale layered SiC5 with spin-one transition', 'Ab initio study of the electron transport in graphene', 'Temperature control of spin-wave transmission in magnetic nanocrystals', 'Thermodynamic studies of the interband current transport in monolayer ZnO', 'On the origin of a conformation in graphene on a hexagonal substrate', 'Electronic Dynamics Prediction of Electron Pronissivity and the Quantum Optical Properties of La5La4Mn3O6', 'Intrinsic and extrinsically modified magnetic order in ferromagnet Li1-xMnxSbxO3001', 'A ferroelectric spin-orbit interaction', 'Temperature dependent nonlinear magnetization in monolayer WS2', 'Ab initio molecular dynamics simulation of the mechanical force of graphene-based thin films', 'Charge-to-Teller transition in Mn-doped Na2MnO3 a first-principles study', 'Temperature dependence of the temperature-dependent conductivity of Fe3Ti3O14 films A First-principles study', 'Influence of Interlayer Defects on Thermoelectric Properties in a Stabilized Fe3NiBi2Te3 Nanoribbon', 'Efficient and effective approach for an arbitrary model model of solid carbon nanotube carbon from first principles', 'Incentrated CsCl2Hs high-temperature phonons', 'Defects in a semiconductor molecule are dependent upon how much the carrier molecule under strain is affected by strain', 'Tuning the electronic state of epitaxial epitaxy by perpendicular ferroelectric-tuned Raman scattering', 'A new method for the accurate quantization of exchange bias calculations in binary phonon sources using a matrix-based linear model for symmetry field conditions', 'Microscopic evidence for structural phase-change and structural magnetism in Kondo semimetals', 'Thermodynamic Characteristics of a Graphene Crystal', 'Coexistence of a phase in semiconducting topological insulator MnBi2O7 A first-principles study', 'Large magnetic field induced ferromagnetic semiconductor transition in SrTiO3 heterojunction', 'Electromagnon Photoconductivity and High-Thermal Conductance in a Superconducting Material', 'High-Pressure Optical Spin Torque in Amorphous Thin Films Topological Insulator', 'Unraveling the phase control of magnetized metals from first principles', 'Pressure-induced phase transition of a doped single-crystalline carbon nanotube', 'High-order spin-orbit coupling in single-photon YBa2SO4', 'Probing the transport properties and electron-hole-phase dynamics of MnFeO3 via molecular-thermal interactions', 'Unsaturable thermal conductance dependence of metallic oxygen in metallic carbon nanotubes', 'A new model for the structure of charged superconductors with quasi-zero and quasi ordered states', 'The Origin of Elastic Field and Dielectric Response of WTe2', 'Inactive Phase Change in the Magnetoelectric Dielectrics of YBaZnPt2S6', 'In-situ mechanical characterization of two-dimensional semiconductor nanocomposites on SbTe', 'Two-dimensional nanocrystal with high thermal conductivity', 'Quantum magnetic disorder and the influence of bismuth and oxygen on the density changes in bulk Au', 'Unraveling the Role of Ferromagnetic Densities in the Stability of Bismuth Oxygen', 'Ab initio study of van der Waals topological insulator phases', 'Spin Hall effect in ultra-thin superconductor XMn3O7-xLi4O3 - a Monte Carlo study', 'Anvil-soft-to-slip solid metallic oxides with 2D soft-tubes', 'Charge Doping in the Mg2O3O100 system and its effects on magnetic and polar fields', 'Temperature-induced phase transition and its impact on the ferroelectric properties of ZnO', 'Maintained superconductivity on non-equilibrium solid-state materials', 'Doping theory of thermal expansion of NiC on TaSe2', 'Magnetic structures and spin-orbit coupling in single crystal Ta2Cu2O6', 'Anomalies and band structures in ultrafast magnetic tunnel junctions with the magnetic interstitial temperature', 'Probing the origin of surface plasticity in polymers with a different composition approach', 'Excited state excitation induced by excited-state coupling in Co2FeO2 a first-principles study', 'A non-linear spin-orbit wave function for monolayer insulator', 'Effect of Temperature on the Pressure Induced by Multiscale Variations', 'Electron mobility and magnetic anisotropy in two-dimensional two layers', 'Substrates of the magnetic anomaly of La07Sr03MnO3 in La067Ca07Mu111 surface', 'A new superconducting topological insulator by molecular quantum excitation', 'Electronic structure of CuO2 thin films', 'Electronic structure and magnetic properties of the FeSnBrO3 thin films on Hg001', 'Spin-orbit torque distribution and magnetism in superconducting SrTiO3', 'A Machine Learning Approach to Determine the Rate of Defect Recovery by Electron Transport']
astro_fake = ['A short time scale model for gamma-ray bursts', 'Relativistic optical radiation emission from ultra-high-energy cosmic rays', 'Two-dimensional modeling for accretion flows', 'A nonlinear gravity model for short-term pulsar mass accretion in a hot AGN', 'The Role of the Rotating Pulsar Magnetar in the Timing Spectrum of Galactic Blazars with Gravitational Waves', 'J16581333A an X-ray pulsar near the heart of an advanced hyperthermal magnetorotational cluster', 'The LMC of a pulsar -- pulsars and high-energy spectra', 'Transients from Gamma Rays From the Black Hole X-ray Binary SSX 339-462', 'The long-term variability of X-ray emission from the blazar ANTARES', 'A spectral look at magnetized relativistic interactions in the galactic nucleus', 'The New Emission Type Ia Supernova Bursts and the Emergence of LIGO-LAT', 'The Evolution of Gamma Ray Emission during the 2005 X-ray Burst', 'The Optical and Spectral Variability of OJ 287 and IV Afterglow', 'Rotation-Driven Neutrino Detection using MHD Observations', 'Unveiling the Progenitors of Massive Binary Jets in Active Galactic Nuclei and the Implication for Gravitational-Wave Detection', 'Accretion-Like Events in a Supernova Remnant', 'General Relativistic Reconnection Relation between Mass and Decay of Spin-Rotating Neutron Stars', 'A comparison of three distinct sources of X-ray emission from the LMCX catalog using a non-stationary detector', 'The Search for Cosmic Ray Transient Excess in the Galactic Center', 'Cosmic ray and gamma rays from Galactic nuclei', 'General Relativity Revisited', 'The GRB 180810O gamma-ray timing variability of the BeX-Ray Source PKS 4349-5124', 'X-ray Luminosity Observations of the GRB Blazar 4R 1854-2215', 'A novel approach to the search for gravitational waves from the galactic center using the XMM-Newton survey', 'A high-energy emission of high gamma rays from black holes', 'Assembling the Gamma-Ray Source GW170817', 'Informing sub-J0843-30 with INTEGRAL', 'SDEA - A Multi-TeV Source in the Galactic Center', 'A Model for Long-Range Supernova Remnants in Blazars', 'The RadioX-ray Flux of GRB 4342-44 in M87', 'Simultaneous optical and radio observations of BL Lacertae', 'The evolution of LIGOVirgo X-ray burst timing variability and spectral consistency', 'GRB 190117A shows the lowest bound on mass in the inner region of the X-ray binary V404', 'Multi-dimensional gamma-ray observations of J1708-64 with NuSTAR', 'X-ray Observations of the Numerical Neutron Star Merger GW170817', 'Spitzer Astrophysics in the Cosmic-Ray Era', 'An improved method for studying the properties of binary black hole companions in the optical and X-ray range of ESO 1048-49', 'The Role of the Quark Core in Black Hole Accretion Disks', 'Positron emission from nonthermal Pulsar Wind Nebulae', 'An updated perspective on the photosphere of black hole merger gamma-ray bursts', 'The magnetic field in the accreting neutron star GW190417', 'The influence of the wind-based microbubbles on the radio-loud pulsar PKS J0536-4543', 'Discovery of a supernova remnant in the Northern Region of NGC 3105', 'A study of the microquasar GW170817', 'Evolution of the X-ray Binaries of Swift Jets and Gravitational Waves', 'The X-ray Variability of the Galactic Center Background Relativistic Radiation Transport', 'The Fermi-LAT Type Iax Supernova Remnant UHECR 4676-097', 'Neutrino-rich matter in accretion discs', 'The Supernova Remnant MHD G071913B and its Role in Super-Enviroment Superheating Stellar Winds', 'X-ray and radio emission from the core of the Type II Neutrinos ULGR J1710-4083', 'Gamma-ray emissions from gamma-rays accretion', 'Three-Dimensional Particle Energy Absorption Simulations of Type II Supernova Remnants', 'Spectral Properties of the Local Progenitor Type III Ultra-High Energy Cosmic Ray Emission from the Fast Radio Burst Active Galactic Nucleus', 'Quasi-periodic oscillations from axion-like neutron star winds', 'A deep redshift and spectral profile of the radio binary FRB 50116A-04', 'A new LRA-driven variability model for the long-term variability of radio radio pulsars', 'High energy emission from the binary binary black hole Mergers', 'The Relativistic Geometry of Dark Matter From Type Ia Supernovae to Black Holes', 'Unobservable Galactic blazar magnetar emission from a wide range of high-energy emission', 'HAWC gamma-ray emission in black holes with low relativistic gravity', 'Neutrino-driven tidal heating events the hard X-ray and gamma-rays of high-energy cosmic rays', 'J1305-4420-4533 a strange globular cluster I an enigmatic low-mass black hole mass hadrons and spin oscillations', 'Signatures of the Electromagnetic Pulse from the Fermi J071417545 with the Galactic Center Telescope', 'Discovery of a high-energy neutron star outburst in GRB 544718A', 'On the origin of magnetars for massive AGN', 'The GRB as a model for the cosmic ray timing variability of SGRB', 'The First New-Time Galaxies with Ejecta', 'IGR J18092162 a dark matter-poor pulsar pulsation model with strong magnetized polar-circuit feedback', 'New perspectives on the magnetospheres of black hole and black-hole binary mergers', 'Particle acceleration for compact objects and their gravitational wave signatures', 'The pulsar magnetic field', 'A fast fast radio burst at the speed of radio bursts', 'The accretion disk-forming blazar SN2017YUVL J1548-6041 is a rare and puzzling out of the 3C17 Blazar catalog', 'The magnetic field of gamma-ray bursts at the Galactic center and sub-CXAS', 'VLBI Observatory and Advanced Photometric Observations of the Crab Nebula', 'Two-Quark Constraints on the Structure of Dark Matter Mass Evolution The Implications of a Hard X-ray Pulsar', 'Interacting with GRB jets in the outer shell of an intergalactic medium', 'Measuring the Interferometry of High Energy Cosmic Rays from Active Galactic Nuclei', 'Dense matter models of supernova evolution', 'NGC 1605 is a gravitational-wave cataloging star with multi-band gamma-ray emission from the first 100 MHz of its era', 'The Gamma-ray burst afterglow from gamma-Ray bursts', 'X-ray Binaries as Evidence of Blazars within Gravitational Wave Detectors', 'General Relativity of the Narrow Line Gravitational Wave Observatory', 'The Radio Evolution of Gamma-ray Pulsar NGC 120519', 'Neutrino Acceleration in Core-collapse Supernova Remnant Supernovae - a Lorentz Principle Approach', 'LISA Search for Cosmic Rays from Active Galactic Nuclei with the Kerr Code', 'A comparison between the radio-star signal rate and the interstellar medium density with TELESR', 'Fast emission from the M87 supernova explosion during the 2011 outburst I Detection of massive jet shells with FRB', 'What we mean by the', 'A new gamma-ray signal from Type Ia supernovae', 'The Progenitor Type IIC AstroSat Mission The Very High Energy Pulsar XMM-Newton Fermi', 'Constraining Relativistic Magnetospheres at High Absorption Rates with Fermi Black Holes', 'Supernova Remnants', 'Ejecta-like Jet Formation in Hydrogen-Rich Quasars', 'Long Gamma-Ray Emission from AGNs', 'On the gamma ray emission from the supernova remnant 1C 5629', 'High-energy gamma-ray emissions from a multi-jet origin-- and timing of the flare', 'What does the inner core of the Quasi-Periodic Magnetorotational Fluid Emission from the Black Hole PKS J1739-5334', 'The XMM-Newton Observing Telescope', 'Cosmic-Ray Line Imprints in Blazar MHDs']
hepEX_fake = ['Long-lived proton-proton fusion particles of Pb-Pb collisions', 'The QCD and the B0-K K-D decays', 'Pomeronium production and its asymmetries in the Higgs Boson', 'A New Method to Detect Spectroscopy and Measure Muon Neutrino Production with Drell-Yan Interaction at HERA', 'Probing the existence of a non-standard Jpsi K from the experiment of pto pbar production in pp collisions at sqrts 196 TeV', 'LSPK-SATD Experiment', 'Lattice Resonance Analysis of N2K Jpsi with Top-Quark-Leading Decays', 'The effect of the relative flow of pion and pions in a finite-like-shape state', 'LHCb Searches with High Performance Higgs Bosons', 'Prospects for WLO-tagged dark matter', 'COSINE II Physics Data - An Update', 'The production of the weak muon gamma in association with a p-Pb collision', 'A Time-Resolved Tracker for LEP Experiments at HEP', 'An AOSPECT QCD study with a deep analysis of the HERA signal', 'Neutrino Mass Measurements at the CAST LHC and NABAR', 'Constraints on the mass of neutrinos in the solar system at atmospheric scales', 'Modeling the distribution of the energy of pPb and the rate of muon recoiling', 'A measurement of the ratio of electroweak symmetry asymmetries to the three-jet cross section in pp collisions at sqrts 8 TeV with the ATLAS detector', 'Measurement of the Time-Dependent CP Violation in the W-Km- Decays of an Electron-Lead Pair', 'Measurement of two-jet production cross sections and correlations within the ee- to the W-boson in e e- annihilation at sqrtsrm NN 502 TeV', 'First results of the ALICE experiment', 'SLD barD piL to SND- and S-barD tau- annihilation', 'MSSM Project T2K Program Summary', 'A new approach to the production and quantization of Dpm and the Z0 sum rules using the HERA Experiment', 'Prospects for new physics on the pion', 'K3 production via the NA68 and SUSY experiments at the LHC', 'New constraints on the electron-flavor states of the sigma pi- decay', 'Measurement of the top quark mass ratio K1psi-K0pi0b in the ATLAS detector', 'The future of the LHC and the future prospects for LEP', 'Determination of the fraction of nucleon pairs in decays to a pair in pp collisions at sqrts 7 TeV', 'Neutrino Backgrounds at Large Hadron Colliders', 'Large scale studies with two-photon scattering', 'Recent Measurement of t-tbar Mixing Mass in High Energy Scattering at sqrts 8 TeV', 'New Results from the Next-to-leading order analysis of the LHC Neutrino Spectroscopy for the B0 - Bpi0 collisions', 'B-meson decays to lepton-flavor and b to dijet production in p-Pb collisions at sqrts 13 TeV', 'The Pairs of the Solar Neutrino Detectors', 'A method of detection of the invisible invisible hadronic parton in nucleon scattering at ALICE', 'Direct measurement of the cross section difference between the proton and the Higgs boson in pp collisions at sqrts 7 TeV', 'Search for Rare Flavored Decays of the CKM to pi- K and its Supersymmetry with the ATLAS Detector', 'Jpsi gamma decay constants for the ATLAS detector', 'Energies of the Photon Inward Coupling', 'Gluon Resonances in the Muon Region from the 3D Search for Heavy Mesons at HERA', 'Higgs-quark pair production from bb meson and tau-etac collisions at sqrts 8 TeV in ee- collisions', 'Study of nuclear reactions in a resonant transverse mass function at sqrts16 TeV', 'Measurement of the Jpsi-jet cross section at sqrts8 TeV', 'One-shot experiment to determine the structure properties of four-jet jets in collisions with strong weak top-quark pairs', 'On Bto lepton decays', 'On neutrino interactions', 'The Probing the B0s0S Mixing Effect with Liquid Argon at DUNE', 'A Higgs boson search at CDF', 'Dijet production of high energy particles in top quark pairs using hadrons', 'A comprehensive evaluation of the H-wave and B- wave measurements of K0-K-pi0p mixing at HERA and the ALICE experiment', 'A new high-energy dipole moment for dark matter searches', 'Pion polarization in pp collisions at sqrts8 TeV', 'A study of quarkonium production in the triple-charm system in pp collisions at sqrts7 TeV using a lattice-like lattica matrix', 'Measurement of the lepton decays FZ2 to ZZZ bosons', 'Observation of the decay Bb0to tbart0pipi rightarrow', 'Longitudinal distribution of nuclei and neutrinos in pp collisions at sqrts 13 TeV', 'New Limits on Unfolded Multitude Fluctuations for New Physics', 'The SADIC-A Low Q2 Phase Detection System of the CLIC Laboratory', 'Baryonic Nucleon Effects on the Upsilon3P Higgs Interactions at the Tevatron', 'SUSY on a W-type structure using the CDF framework', 'Heavy-Quarkonium Production at BESIII', 'Probing the origin of the proton by a gamma-ray background as a probe of S-particle coupling', 'Higgs mass fluctuations and constraints', 'The nature of the dark jet properties', 'Physics of a Supersymmetric Baryon', 'A new form factor for the CMS-based photon detector', 'Scaling up B to pi pi with the CDF project', 'Measuring the effective energy and its dependence on the flow parameters in a light-gammagamma-dynamically tautomeric dipole moment', 'Status of the T2S Simulation Facility', 'New study of D0 - kDKbar measurement in Pb-Pb collisions at sqrts 18 TeV', 'The Measurement of the Jpsi Charm from X3872', 'Large-X Physics with CP Violation', 'An Enhanced Search for Double Beta Decay for Hadrons', 'A novel method for calculating the cross section on the quark from KLOE', 'Search for neutrinoless double beta decay in D0 to pp collisions at sqrts 13 TeV with the ATLAS detector', 'Search for tau lepton production in pp collisions at sqrts 13 TeV using ATLAS and the muons', 'Heavy neutrino oscillations from small isolated particles', 'Direct measurement of phi -- mu and its lifetime at sqrts196 TeV', 'The CMS for Higgs-like decays', 'Recent Results on the Neutrino Resonance', 'First measurement of the lepton shape in ATLAS data', 'Lepton Pair Production at LEP', 'A New Method for High-energy Determination of Neutrino-Proton Mass using Deep Neural Networks', 'Measurement of W and Z - production cross sections in pp collisions at sqrts 13 TeV', 'New approaches to B to Jpsi production in B-s to pi pi decays', 'Three-Luminous CP Asymmetries in Photon Couplings', 'Search for a weak B-Keto', 'New experiments and a new model for particle physics search', 'Asymmetry of Ds to p-ds at sqrts13 TeV', 'Inclusive CP violation and the decay ee- to CDF e- and pbar pp collisions', 'Topological and Direct CP Violation of Dark Matter', 'The Search for Heavy Q2 and Supersymmetry in the Search of the Energy Higgs Boson', 'The Higgs boson at HERA', 'A novel probe of the electroweak coupling constants of protons and neutrinos', 'Higgs-photon interactions with a neutrino near neutrinoless Doublebeta and proton', 'The impact of longitudinal perturbations', 'The JPC-BHC-VASU experiment', 'Evidence for a Higgs Boson at the LHC using the Lambdab2-Kho-mu-tagged nucleon']



# Pull data according to user selection
if   category == 'Astrophysics':
  selected_real = astro_real
  selected_fake = astro_fake

elif category == 'Condensed Matter':
  selected_real = condensed_real
  selected_fake = condensed_fake

elif category == 'High Energy Physics - Experiment':
  selected_real = hepEX_real
  selected_fake = hepEX_fake

elif category == 'Tissues and Organs':
  selected_real = tissue_real
  selected_fake = tissue_fake

print('You Selected: ' + str(category))

# Generated titles often have a capital first letter only
selected_real = [title.lower() for title in selected_real]
selected_fake = [title.lower() for title in selected_fake]

import ipywidgets as widgets
from IPython.display import display
import numpy as np

class buttons:
  def __init__(self, selected_real, selected_fake):

    # Create Buttons
    self.button_A = widgets.Button(description="A is Real")
    self.button_B = widgets.Button(description="B is Real")

    # Printing Text
    self.output = widgets.Output()

    # Store real and fake titles
    self.selected_real = selected_real
    self.selected_fake = selected_fake

    # Expands title text boxes so everything will fit
    self.text_layout = widgets.Layout(width='90%')

    # Note: Pick two. With all three, one is redundant
    self.counter = 0
    self.correct = 0
    self.incorrect = 0

    # Create a random ordering for the real and fake titles
    self.real_indices = np.arange(len(selected_real))
    self.fake_indices = np.arange(len(selected_fake))
    # Shuffling is in place
    np.random.shuffle(self.real_indices)
    np.random.shuffle(self.fake_indices)

    # Set up the first round
    self.pull_real_fake()
    self.display_interface()

  def display_interface(self):
    left_box = widgets.VBox([self.button_A, self.button_B])
    right_box = widgets.VBox([self.string_A, self.string_B], 
                             layout=self.text_layout)
    interface = widgets.HBox([left_box, right_box], layout=self.text_layout)

    display(interface)

  def pull_real_fake(self):

    # Select a single real and fake title according to shuffled indices
    self.real = self.selected_real[self.real_indices[self.counter]]
    self.fake = self.selected_fake[self.fake_indices[self.counter]]

    self.counter += 1

    # Determine display order of titles
    if np.random.randint(2) == 0:
      self.string_A = widgets.Text(self.real, layout = self.text_layout)
      self.string_B = widgets.Text(self.fake, layout = self.text_layout)
      self.real_button = 'A'
    else:
      self.string_A = widgets.Text(self.fake, layout = self.text_layout)
      self.string_B = widgets.Text(self.real, layout = self.text_layout)
      self.real_button = 'B'

  def on_button_clicked_A(self, b):

    if self.real_button == 'A':
      self.correct += 1
      correctness = 'Correct! The real title was: {}'
    else:
      self.incorrect += 1
      correctness = 'Oops! The real title was: {}'

    # Display the message within the output widget.
    print(correctness.format(self.real))
    print('Accuracy: {}/{} = {}'.format(self.correct, 
                                        self.correct + self.incorrect, 
                                        self.correct/(self.correct + self.incorrect)))
    # Some white space
    print()
    print()
    
    # Set up next round
    self.pull_real_fake()
    self.display_interface()

  def on_button_clicked_B(self, b):

    if self.real_button == 'B':
      self.correct += 1
      correctness = 'Correct! The real title was: {}'
    else:
      self.incorrect += 1
      correctness = 'Oops! The real title was: {}'

    # Display the message within the output widget.
    print(correctness.format(self.real))
    print('Accuracy: {}/{} = {}'.format(self.correct, 
                                        self.correct + self.incorrect, 
                                        self.correct/(self.correct + self.incorrect)))
    # Some white space
    print()
    print()
    
    # Set up next round
    self.pull_real_fake()
    self.display_interface()

# Begin the game
button_object = buttons(selected_real, selected_fake)
button_object.button_A.on_click(button_object.on_button_clicked_A)
button_object.button_B.on_click(button_object.on_button_clicked_B)
```
